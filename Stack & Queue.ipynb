```python
class Node:

    def __init__(self,data):
        self.data = data
        self.next = None
     
class Stack:
    def __init__(self):
        self.head = None
     
    # Case(a)        Checks if stack is empty
    def isempty(self):
        if self.head == None:
            return True
        else:
            return False

    # Case(b)        add data to stack(at begining)
    def push(self,data):
        if self.head == None:
            self.head=Node(data)
             
        else:
            newnode = Node(data)
            newnode.next = self.head
            self.head = newnode
    # Case(c) Delete data from the stack.
    def pop(self):
         
        if self.isempty():
            return None
             
        else:
            # Removes the head node and makes
            #the preceeding one the new head
            poppednode = self.head
            self.head = self.head.next
            poppednode.next = None
            return poppednode.data
     
    # Returns the head node data
    def peek(self):
         
        if self.isempty():
            return None
             
        else:
            return self.head.data
     
    # Case(d) Display elements of stack   
    def display(self):
         
        iternode = self.head
        if self.isempty():
            print("Stack Underflow")
         
        else:
             
            while(iternode != None):
                 
                print(iternode.data,"->",end = " ")
                iternode = iternode.next
            return
    def __del__(self):
      print("Destructor called")


S1 = Stack()
print(S1.isempty())
S1.push(1)
S1.push(2)
# Case(b)
S1.push(3)
S1.push(4)
print("\n")
# Case(d)
S1.display()
print("\n")
 
# Case(c)
S1.pop()
S1.display()
 
print("\n")
# Case(e)
del S1
```

```python
class Rectangle:
  def __init__(self):
    self.l = 4
    self.b = 3
    #Default Constructor
  def def_rectangle(self):
    print(f"Area = {self.l * self.b}")
    #Parametrized Constructor
  def para_rectangle(self, l, b):
    print(f"Area = {l*b}")
r1 = Rectangle()
r1.def_rectangle()
r2 = Rectangle()
r2.para_rectangle(11,2)
```

```python
class Queue(object):
  #constructor
    def __init__(self, size):
        self.queue = []
        self.size = size

    def __str__(self):
        myString = ' '.join(str(i) for i in self.queue)
        return myString

    #adding elements(enqueue)
    def enqueue(self, item):
        if(self.isFull() != True):
            self.queue.insert(0, item)
        else:
            print('Queue is Full!')

   #removing elements(dequeue)
    def dequeue(self):
        if(self.isEmpty() != True):
            return self.queue.pop()
        else:
            print('Queue is Empty!')

      # to check if queue is empty
    def isEmpty(self):
        " Queue is empty "
        return self.queue == []

      # to check if queue is full
    def isFull(self):
        " Queue is full "
        return len(self.queue) == self.size

    def __del__(self):
      print("Destructor called")

#defining a queue of size 10
    q1 = Queue(10)
    q1.enqueue(4)
    q1.enqueue(5)
    q1.enqueue(6)
    
    print(q1)
    
    q1.enqueue(1)
    q1.enqueue(2)
    q1.enqueue(3)
    
    print(q1)
    
    q1.dequeue()
    
    print(q1)
    
    del q1
```
