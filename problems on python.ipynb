```python
#Problem-3
class Team:
  def __init__(self, name, salary):
    self.name = name
    self.salary = salary
 
  def increment(self, amount):
    self.salary += amount

  def display(self):
    print(f"Team name: {self.name} \nTotal Salary: {self.salary} \nIncremented salary: {self.salary}")

class Player:
  def __init__(self, name, salary, iplteam):
    Team.__init__(self, name, salary)
    self.iplteam = iplteam
    self.new_amount = 0
  
  def prize(self, amount, bonus):
    new_amount = amount * bonus
    Team.increment(self,new_amount)

  def display(self):
    print(f"Name:  {self.name} \nIPL Team: {self.iplteam} \nSalary: {self.salary} \n")

t1 = Team("PWI", 300000)
t1.increment(10000)
t1.display()
p1 = Player("Yuvraj", 10000, "PWI")
p1.prize(5000, 2)
p1.display()
```

```python
# Problem 4: Alternative Constructor 
class MyDate:
    def __init__(self, year, month, day):     
      self.year, self.month, self.day = year, month, day    
    @classmethod
    def from_str(cls, datestr):
        parts = datestr.split("-")
        year, month, day = int(parts[0]), int(parts[1]), int(parts[2])
        return cls(year, month, day)       # Return the class instance 
    def display(self):
      print(f"Your b'day detais are: \n1.Year: {self.year} \n2.Month: {self.month} \n3.Day: {self.day}") 
bd = MyDate.from_str('2002-11-18')   
bd.display()
```

```python
#Problem-5
class Name:
  def __init__(self, first_name, middle_name, last_name):
    self.first_name = first_name
    self.middle_name = middle_name
    self.last_name = last_name
    print(f"First Name: {first_name} \nMiddle Name: {middle_name} \nLast Name: {last_name}")
  @classmethod
  def break_str(cls, name):
    parts = name.split(" ")
    first_name, middle_name, last_name = parts[0], parts[1], parts[2]
    return cls(first_name, middle_name, last_name)
p1 = Name.break_str("Bhanwar Yuvraj Singh")
```
